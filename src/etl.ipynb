## Importação e Carregamento dos Dados

Importação da biblioteca Pandas e carregamento do dataset para realização da análise inicial da estrutura e tipagem dos dados.

```python
import pandas as pd
df = pd.read_csv("../data/spotify.csv")
df.head()
```

## Validação da Estrutura e Tipagem dos Dados

Análise das informações estruturais do dataset utilizando `df.info()` para verificação da tipagem das colunas, quantidade de registros não nulos e consistência dos dados.

Foi identificado que o dataset não possui valores nulos, considerando que todas as colunas apresentaram a mesma quantidade de valores não nulos em relação ao total de registros.

Também foi verificado que a coluna `signup_date` estava definida como string, sendo necessária a conversão para o tipo datetime para adequação das análises temporais.

Além disso, a coluna `inactive_3_months_flag` foi identificada como inteiro (`int64`), embora sua definição no dataset represente um valor booleano, exigindo a conversão para o tipo `bool`.

```python
df.info()
```

```python
df["signup_date"] = pd.to_datetime(df["signup_date"])

df["inactive_3_months_flag"] = df["inactive_3_months_flag"].astype(bool)
```

```python
df.info()
```

## Verificação de Duplicidade de Registros

Foi realizada a análise de duplicidade de registros com base na coluna `user_id`, utilizando o método `duplicated().sum()` para identificar a existência de possíveis entradas repetidas no dataset.

O resultado retornado foi 0, indicando que não foram identificados registros duplicados.

Dessa forma, conclui-se que a chave identificadora `user_id` possui unicidade garantida, assegurando a integridade dos registros e evitando impactos em métricas agregadas e análises estatísticas subsequentes.

```python
df["user_id"].duplicated().sum()
```
